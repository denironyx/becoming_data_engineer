```python
## Waiting until thingss break
per_page=250
page = 1
json_df = []

while True:
    print("----")
    url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&per_page={per_page}&page={page}"
    print("Requesting", url)
    response = requests.get(url)
    data = response.json()
    
    # Debug by printing out the first pokemon
    try:
        print(data[0])
    except:
        print("There was an error!")
        break

    ## if we did find pokemon, add them 
    ## to our list and then move on to the next
    json_df.extend(data)
    page = page + 1   
```

```python
df_coin_market = pd.DataFrame(json_df)
df_coin_market_filter = pd.DataFrame(json_df, columns=['id', 'name', 'current_price', 'market_cap', 'high_24h', 'last_updated'])
df_coin_market
```
